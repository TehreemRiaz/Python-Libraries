# 2. Generate and plot data: (2)

- y = [3, 5, 7, 2, 1]

### Plot the data as a scatter plot

```python
df = pd.DataFrame({"x":[1, 2, 3, 4, 5], "y":[3, 5, 7, 2, 1]})
df
plt.scatter(df['x'], df['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot')
# plt.savefig('scatterplot.png')

plt.show()
```

# 6. Subplots and multiple plots: (4)
Create 2 Subplots
+ Bar Plot
+ Scatter Plot

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.bar(df['x'], df['y'], color='skyblue', edgecolor='black')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('X Bar Plot')
ax1.grid(True)

# Scatter plot
ax2.scatter(df['x'], df['y'], marker='s', color='red')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Scatterplot of x and y')
ax2.grid(True)
plt.savefig('scatterplot.png')
plt.tight_layout()

plt.show()
```

# 9. Layout and composition: (4)

Explore the impact of different plot sizes and aspect ratios by creating a small square plot and a larger wide plot using the 'scatter' function.
- Small = 5x5
- Large = 10x5

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(5, 5))
plt.scatter(x, y)
plt.title('Small Square Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.scatter(x, y)
plt.title('Large Wide Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

Modify the legend placement to the upper left and adjust the positioning of the annotation for the point (2, 5) in a scatter plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 4, 6]

plt.scatter(x, y, label='Data Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot with Annotation and Legend')

plt.annotate('Point (2, 5)', xy=(2, 5), xytext=(3, 6),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.legend(loc='upper left')

plt.grid(True) 
plt.show()
```

Enhance the use of whitespace in a plot by adjusting the spacing around the plot area and creating a scatter plot with improved whitespace.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(8, 6)) 
plt.scatter(x, y)

plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

plt.title('Scatter Plot with Improved Whitespace')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

plt.show()
```

# 11. Customize plot styles: (4)

Experiment with line styles, marker shapes, and color gradients:
- Dashed Line
- Triangle Marker (Green)

Utilize color gradients to represent data using a specified colormap, for example, 'YlGnBu', to create a scatter plot with a color bar.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating sample data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.rand(100) * 100  # Sample data for color gradients

# Plotting
fig, ax = plt.subplots()

sc = ax.scatter(x, y, c=z, cmap='YlGnBu', alpha=0.8)

cbar = plt.colorbar(sc)
cbar.set_label('Color Intensity')

ax.plot(np.sort(x), np.sort(y), linestyle='--', color='black')

ax.plot(x, y, marker='^', linestyle='None', color='green')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Scatter Plot with Line and Marker Styles')

plt.show()
```
